```python
import pandas as pd
import hashlib as hl
```

```python
athlete_regions = pd.read_csv("data/athlete_regions.csv")
athlete_regions.head()
```

```python
canada = athlete_regions[athlete_regions["region"] == "Canada"]
canada.to_csv("data/canada.csv")
canada.head()
```

```python
canada.columns
```

```python
canada["Name"].head()
```

```python
hashes = canada["Name"].apply(lambda x: hl.sha256(x.encode()).hexdigest())
hashes.head()
```

```python
canada.insert(2, "HashName", hashes)
canada.head()
```

```python
canada.to_csv("data/canada.csv")
```
